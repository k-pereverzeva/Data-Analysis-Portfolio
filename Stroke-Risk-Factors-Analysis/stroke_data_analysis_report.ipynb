# Analysis of Stroke Risk Factors in Patients: A Statistical Study

## About the Project

This project involves a comprehensive statistical analysis of various factors influencing the risk of stroke among patients. The dataset used for this analysis includes patient information such as age, gender, hypertension, heart disease, body mass index (BMI), smoking status, work type, residence type, and glucose level. By examining these factors, the study aims to identify significant correlations and determine the impact of each factor on the likelihood of a stroke.

## Project Goals

The objective of this project is to identify and analyze key risk factors associated with stroke occurrences in patients. Specifically, the goals are to:

* Determine the prevalence of stroke among different demographic groups.
* Assess the impact of age, hypertension, heart disease, and other health indicators on the risk of stroke.
* Evaluate the influence of lifestyle factors such as smoking and employment type on stroke incidence.
* Perform statistical tests to establish the significance of observed correlations.
* Provide insights and recommendations for stroke prevention based on the analysis findings.

## Data Description and Validation

The dataset contains **5110 rows and 12 columns** before cleaning and validation. Each column represents different attributes of patients and their health-related factors. Below is a brief description and validation process for each column:

* **id**: Unique identifier for each patient; 5110 unique values, no cleaning needed.
* **gender**: Gender of the patient; values include 'Male', 'Female', and 'Other'. Only one row had 'Other' gender, so it was removed, leaving 5109 rows.
* **age**: Age of the patient; values range from 0 to 82 years, converted to int64, no cleaning needed.
* **hypertension**: Indicates if the patient has hypertension (0 = No, 1 = Yes); binary values, no cleaning needed.
* **heart_disease**: Indicates if the patient has heart disease (0 = No, 1 = Yes); binary values, no cleaning needed.
* **ever_married**: Indicates if the patient has ever been married ('Yes' or 'No'); no cleaning needed.
* **work_type**: Type of occupation; values include 'Private', 'Self-employed', 'Govt_job', 'children', 'Never_worked', no cleaning needed.
* **Residence_type**: Type of residence ('Urban' or 'Rural'); no cleaning needed.
* **avg_glucose_level**: Average glucose level in the blood; outliers were removed using the Z-score method, but some outliers remain because they fall within plausible medical ranges and are important for analysis.
* **bmi**: Body Mass Index; 201 missing values filled with the median BMI for each gender and age group. Outliers were removed using the Z-score method, but some outliers remain because they fall within plausible medical ranges and are important for analysis.
* **smoking_status**: Smoking status ('formerly smoked', 'never smoked', 'smokes', 'Unknown'); no cleaning needed.
* **stroke**: Indicates if the patient has had a stroke (0 = No, 1 = Yes); binary values, no cleaning needed.
* **age_group** (created during data processing): 7 age groups created for analysis (0-2, 3-12, 13-18, 19-30, 31-45, 46-60, 61+); correctly assigned based on age, no cleaning needed.

After validation and cleaning, the dataset contains **5001 rows and 13 columns** with no missing values. The cleaning process included handling missing values in the BMI column, removing outliers, and ensuring consistency in the data. This cleaned dataset is now ready for further analysis.

**Age:**

**BMI:**
* Shapiro-Wilk Test: p-value = 1.468e-22
* D'Agostino and Pearson Test: p-value = 8.800e-37

Non-parametric statistical tests will therefore be more appropriate for further analysis of age, average glucose level, and BMI.

## How Different Metrics Influence Stroke Incidence

### Categorical Data

The impact of various factors on stroke incidence was examined using Chi-Square tests for categorical data, revealing significant associations as follows:

- **Gender**: Male individuals exhibit a slightly higher incidence of stroke compared to females, but this difference was not statistically significant (Chi-Square = 0.030, p-value = 0.863). Therefore, we accept the null hypothesis that gender does not have a significant effect on stroke incidence.
  
- **Hypertension**: Presence of hypertension significantly increases the likelihood of experiencing a stroke compared to absence of hypertension (Chi-Square = 71.113, p-value < 0.001). This result is statistically significant, indicating that hypertension is associated with higher stroke incidence.

- **Heart Disease**: Having heart disease markedly elevates the risk of stroke compared to individuals without heart disease (Chi-Square = 77.421, p-value < 0.001). This finding is statistically significant, suggesting a strong association between heart disease and stroke incidence.

- **Smoking Status**: Former and current smokers demonstrate higher stroke incidence rates compared to those who have never smoked or have an unknown smoking status (Chi-Square = 26.904, p-value = 0.00001). This result is statistically significant, indicating that smoking status influences stroke risk.

- **Work Type**: Self-employed individuals and those in government jobs show higher stroke incidence rates compared to those in private sector roles or children (Chi-Square = 50.250, p-value < 0.001). This finding is statistically significant, suggesting an association between work type and stroke incidence.

- **Residence Type**: Urban residents have a slightly higher incidence of stroke compared to rural residents, but this difference was not statistically significant (Chi-Square = 1.429, p-value = 0.232). Therefore, we accept the null hypothesis that residence type does not significantly impact stroke incidence.

- **Age Group**: Older age groups, particularly those aged 61 and above, exhibit markedly higher rates of stroke incidence compared to younger age groups (Chi-Square = 312.721, p-value < 0.001). This result is highly statistically significant, indicating a robust association between age group and stroke incidence.

These findings underscore the importance of addressing modifiable risk factors such as hypertension, heart disease, smoking status, work type, and age group to mitigate stroke risk effectively through targeted preventive measures and interventions.

### Numerical Data

The data indicates a clear difference in average age between individuals who have had a stroke (67.65 years) and those who have not (41.74 years), underscoring age as a significant factor associated with stroke incidence. Therefore, I have decided to conduct further tests using data that includes only the high-risk age groups, specifically those with stroke rates greater than 1%.

Avg age by stroke

Based on **the Mann-Whitney U Test** results for patients in the high-risk age group:

- The **Average Glucose Level** for patients without stroke is 112.47, whereas for patients with stroke it is 129.90. The Mann-Whitney U Test yielded a statistic of U = 285699.0 with a p-value of **0.000197**, indicating a statistically significant difference in average glucose levels between patients with and without stroke.

- In contrast, the average **BMI** for patients without stroke is 30.36, and for patients with stroke it is 29.96. The Mann-Whitney U Test resulted in U = 238731.0 with a p-value of **0.331**, suggesting that there is no statistically significant difference in BMI between patients with and without stroke.

These findings suggest that average glucose levels may be more strongly associated with stroke incidence in the high-risk age group compared to BMI. Further analysis and consideration of glucose levels as a risk factor for stroke could be warranted based on these results.
