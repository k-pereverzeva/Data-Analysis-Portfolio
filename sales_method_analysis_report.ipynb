# New Product Sales Methods Analysis

## About the Project

Six weeks ago, we launched a new line of office stationery, and we tested three different sales strategies - targeted email, phone calls, and a combination of both.

## Business Goals

As these methods vary in their demand on the team's time, our business goal was to determine the differences between them and to identify the most effective sales approach that we should continue using.

This report provides an analysis of the given data and presents my thoughts on choosing a sales method. It summarizes the results obtained from the analysis and provides recommendations for the company.

## Business Metrics

In order to determine the most effective approach, I suggest using **the ratio of the average price of products sold in the last 6 weeks to the cost of each approach** as our metric. While we don't have specific numbers for the amount of time spent on each customer for each approach, we can make some assumptions based on the information provided. It appears that the most time-consuming method is the call approach, as it takes an average of 30 minutes per customer. The email and call approach is likely the next most time-consuming, with email requiring relatively little work from the team and the call taking around 10 minutes per customer. Finally, the least time-consuming approach is likely the email method. 

Based on our data, we can see that the call method, which takes the most time with an average of 30 minutes per customer, generates an average price of 5`$`. On the other hand, the email method, which requires the least amount of time, generates an average price of 10`$`. The email and call method, which falls somewhere in between in terms of time spent, generates an average price of 15`$`. Depending on the resources available to our company, the most effective approach can be chosen based on this metric.

2. **Specific products were bought using different approaches**

	If resources are available, the **email and call strategy** may also be effective as it generated the highest revenue per item sold.
